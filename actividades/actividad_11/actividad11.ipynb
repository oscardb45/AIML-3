### *** Limpieza ***

```python
data = data.drop('customerID', axis=1)
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data.fillna({'TotalCharges': data['TotalCharges'].median()}, inplace=True)
```

### ***Codificar y estandarizar ***

```python
#Convierte las variables categoricas
data['Churn'] = data['Churn'].map({'Yes': 1, 'No': 0})
data = pd.get_dummies(data, drop_first=True)
```

```python
#Escalamos las variables numericas y balanceamos
X = data.drop('Churn', axis=1)
y = data['Churn']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_scaled, y)
```

```python
### Separar datos de prueba y entrenamiento
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

### *** Entrenamos modelo ***

##### Random Forest: tiene buena precisión y es más robusto que un solo árbol

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

### *** Evaluamos modelo ***

```python
y_pred = model.predict(X_test)
```

```python
print("\n--- Resultados ---")

print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, model.predict_proba(X_test)[:,1]))
```

```python
##### Matriz de confusion
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
plt.title("Matriz de Confusión")
plt.show()
```

```python
# Curva ROC
RocCurveDisplay.from_estimator(model, X_test, y_test)
plt.title("Curva ROC")
plt.show()
```

### *** Ajuste de hiperparámetros ***

```python
param_grid = {
    'n_estimators': [100, 200], # Número de árboles en el bosque
    'max_depth': [None, 10, 20], # Profundidad máxima de los árboles (None = sin límite)
    'min_samples_split': [2, 5], # Número mínimo de muestras para dividir un nodo
    'min_samples_leaf': [1, 2] # Mínimo de muestras en una hoja
}
```

```python
# Ajuste de hiperparámetros
grid = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3, scoring='f1', n_jobs=-1)
grid.fit(X_train, y_train)
# Entrenar con los datos de entrenamiento
grid.fit(X_train, y_train)
```

```python
print("Mejores parámetros:", grid.best_params_)
```

### *** Evaluar mejor modelo ***

```python
best_model = grid.best_estimator_
y_pred_best = best_model.predict(X_test)
```

```python
print("Classification Report (modelo optimizado):\n", classification_report(y_test, y_pred_best))
print("ROC AUC Score (optimizado):", roc_auc_score(y_test, best_model.predict_proba(X_test)[:, 1]))
```

```python
# Curva ROC del modelo optimizado
RocCurveDisplay.from_estimator(best_model, X_test, y_test)
plt.title("Curva ROC - Modelo Optimizado")
plt.show()
```
